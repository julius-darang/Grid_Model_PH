```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
# Load the Philippine shapefile from the data folder
phl_map = gpd.read_file("data/PH_Adm2_ProvDists.shp.zip")

# Plot the Philippines map
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
phl_map.plot(ax=ax, color='lightblue', edgecolor='black')
ax.set_title('Philippines - Provinces and Districts')
ax.set_axis_off()
plt.tight_layout()
plt.show()
```

```python
# Load the Philippine shapefile from the data folder
phl_map = gpd.read_file("data/PH_Adm2_ProvDists.shp.zip")

# Load the administrative regions data
admin_regions_df = pd.read_csv("data/admin_regions.csv")

# Filter for Visayas regions based on 'custom_TDP1' column
visayas_admin_data = admin_regions_df[admin_regions_df['custom_TDP1'] == 'Visayas']

# Get the list of province names (adm2_en) for Visayas
visayas_province_names = visayas_admin_data['adm2_en'].tolist()
print("Visayas Province Names:", visayas_province_names)

# Filter the GeoDataFrame to include only Visayas provinces
# Using 'adm2_en' as identified from phl_map.columns
visayas_map = phl_map[phl_map['adm2_en'].isin(visayas_province_names)]

# Plot the Visayas map
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
visayas_map.plot(ax=ax, color='lightblue', edgecolor='black')
ax.set_title('Visayas Region Provinces and Districts')
ax.set_axis_off()
plt.show()
```
